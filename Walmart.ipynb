```python
import matplotlib.pyplot as plt
import seaborn as sns

# Weekly Sales Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Weekly_Sales'], kde=True, bins=30, color='blue')
plt.title('Distribution of Weekly Sales', fontsize=16)
plt.xlabel('Weekly Sales', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# Temperature Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Temperature'], kde=True, bins=30, color='orange')
plt.title('Distribution of Temperature', fontsize=16)
plt.xlabel('Temperature', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# Unemployment Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Unemployment'], kde=True, bins=30, color='green')
plt.title('Distribution of Unemployment', fontsize=16)
plt.xlabel('Unemployment Rate', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Convert Temperature from Fahrenheit to Celsius
df['Temperature_Celsius'] = (df['Temperature'] - 32) * 5/9

# Scatter plot: Temperature (Celsius) vs Weekly Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature_Celsius', y='Weekly_Sales', data=df, hue='Holiday_Flag', palette='coolwarm')
plt.title('Temperature (Celsius) vs Weekly Sales', fontsize=16)
plt.xlabel('Temperature (C)', fontsize=12)
plt.ylabel('Weekly Sales', fontsize=12)
plt.legend(title='Holiday Week', loc='upper right')
plt.show()
```

X-axis (Temperature in Celsius):
The temperature range is from -20C to 40C, with the majority of data points clustered between -10C and 38C. This suggests that most of the data is centered around relatively moderate temperatures, with fewer extreme values on either end.
2. Y-axis (Weekly Sales):
The sales range from 0.5 to 3.5 million, with a concentration between 0.3 million and 3.0 million. This suggests that sales are generally within this range for most of the weeks.
3. Clustered Data Points:
 the data is highly clustered between -10C and 38C, where sales are mostly between 0.3 and 3.0 million. This indicates that the bulk of sales happen in this temperature range, suggesting a relatively stable relationship between temperature and sales.
4. Holiday Flag:
The Holiday_Flag is dominated by blue, which likely indicates non-holiday weeks, with a few orange points representing holiday weeks. This suggests that most sales happen during non-holiday weeks, but holiday weeks still show some presence in the scatter plot.
5. Outliers:
 outliers where sales exceeded 3.5 million when the temperature ranged from -8C to 15C. These outliers could represent specific events or promotions that led to higher sales despite moderate temperatures.
6. General Trend:
The plot suggests that sales are generally highest between -10C and 38C. This range could be considered the optimal temperature for sales, as most data points fall within this range, and sales are concentrated there. There doesnt seem to be a strong positive or negative correlation with temperature, but the clustering suggests that moderate temperatures might be more favorable for sales.

```python
# Calculate the correlation between Unemployment and Weekly_Sales
correlation = df['Unemployment'].corr(df['Weekly_Sales'])
print(f"Correlation between Unemployment and Weekly Sales: {correlation}")

# Scatter plot: Unemployment vs Weekly Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Unemployment', y='Weekly_Sales', data=df, hue='Holiday_Flag', palette='coolwarm')
plt.title('Unemployment Rate vs Weekly Sales', fontsize=16)
plt.xlabel('Unemployment Rate', fontsize=12)
plt.ylabel('Weekly Sales', fontsize=12)
plt.show()
```

The correlation between Unemployment and Weekly Sales is approximately -0.11, which indicates a weak negative correlation. This weak correlation implies that unemployment may not be a major factor affecting sales, or other factors could be influencing the sales more significantly.

```python
# Group by Store and calculate the mean unemployment rate and weekly sales
store_sales_unemployment = df.groupby('Store').agg(
    avg_unemployment=('Unemployment', 'mean'),
    avg_weekly_sales=('Weekly_Sales', 'mean')
).reset_index()

# Sort stores by average unemployment rate to see which ones have the highest unemployment
sorted_store_sales = store_sales_unemployment.sort_values(by='avg_unemployment', ascending=False)

# Display the top stores with the highest unemployment and their weekly sales
print(sorted_store_sales.head())
```

Although the correlation between unemployment and sales is weak, we can see that certain stores with higher unemployment rates are experiencing reduced sales. However, other factors (e.g., store location, local economic conditions, promotions, etc.) could also play a role in these variations.

```python

df['Date'] = pd.to_datetime(df['Date'])

# Extract Month and Year from Date for seasonal trend analysis
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Group by Month and calculate average Weekly Sales
monthly_sales = df.groupby('Month').agg(avg_monthly_sales=('Weekly_Sales', 'mean')).reset_index()

# Plot the trend of average monthly sales
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='avg_monthly_sales', marker='o')
plt.title('Average Monthly Sales Trend', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Weekly Sales', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

There is a clear seasonal trend in the data, with December being the peak month for sales, likely due to the holiday season. Other months like November and June also see notable rises, which could be linked to early holiday sales, summer shopping, or promotional events.

```python
# Calculate correlation between CPI and Weekly Sales
cpi_sales_correlation = df['CPI'].corr(df['Weekly_Sales'])
print(f"Correlation between CPI and Weekly Sales: {cpi_sales_correlation}")

# Scatter plot: CPI vs Weekly Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='CPI', y='Weekly_Sales', data=df, hue='Holiday_Flag', palette='coolwarm')
plt.title('CPI vs Weekly Sales', fontsize=16)
plt.xlabel('Consumer Price Index (CPI)', fontsize=12)
plt.ylabel('Weekly Sales', fontsize=12)
plt.grid(True)
plt.show()
```

The weak negative correlation suggests that CPI doesnt have a major impact on weekly sales in this dataset. There might be other factors influencing sales more strongly than CPI, such as seasonality, holiday effects, or local economic conditions.

```python
# Filter data for Store 5
store_5_data = df[df['Store'] == 5].copy()

# Ensure the 'Date' column is in datetime format
store_5_data['Date'] = pd.to_datetime(store_5_data['Date'])

# Set 'Date' as the index
store_5_data.set_index('Date', inplace=True)

# Plot the time series of Weekly_Sales for Store 5
plt.figure(figsize=(10, 6))
plt.plot(store_5_data['Weekly_Sales'], label='Weekly Sales')
plt.title('Weekly Sales for Store 5')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test on the 'Weekly_Sales' series
adf_result = adfuller(store_5_data['Weekly_Sales'])

# Extract the ADF statistic and p-value
adf_statistic = adf_result[0]
adf_p_value = adf_result[1]

# Print the results
print(f'ADF Statistic: {adf_statistic}')
print(f'p-value: {adf_p_value}')

# Interpret the result
if adf_p_value < 0.05:
    print("The series is stationary.")
else:
    print("The series is not stationary.")
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define the SARIMA model
# For simplicity, we will use (p=1, d=0, q=1) for ARIMA and (P=1, D=0, Q=1, S=52) for seasonal components
sarima_model = SARIMAX(store_5_data['Weekly_Sales'],
                       order=(1, 0, 1),  # ARIMA order (p, d, q)
                       seasonal_order=(1, 0, 1, 52),  # Seasonal order (P, D, Q, S)
                       enforce_stationarity=False,
                       enforce_invertibility=False)

# Fit the model
sarima_results = sarima_model.fit()

# Print the summary of the model
print(sarima_results.summary())
```

```python
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(store_30_data['Weekly_Sales'])
print(f"ADF Statistic: {adf_result[0]}")
print(f"p-value: {adf_result[1]}")

if adf_result[1] < 0.05:
    print("The series is stationary.")
else:
    print("The series is not stationary and may need differencing.")
```

```python
# Apply differencing to make the series stationary
store_30_data.loc[:, 'Weekly_Sales_diff'] = store_30_data['Weekly_Sales'] - store_30_data['Weekly_Sales'].shift(1)

# Drop the NaN values that result from differencing
store_30_data.dropna(inplace=True)

# Check stationarity again on the differenced series
adf_test_diff = adfuller(store_30_data['Weekly_Sales_diff'])
print(f"ADF Statistic After Differencing: {adf_test_diff[0]}")
print(f"p-value After Differencing: {adf_test_diff[1]}")
```

```python
# Step 7: Fit SARIMAX model (without setting Date as the index)
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Use the exogenous variables
exog_vars = store_30_data[['Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']]

# Fit the SARIMAX model (order and seasonal_order can be tuned as needed)
sarimax_model = SARIMAX(store_30_data['Weekly_Sales'], exog=exog_vars, order=(1, 0, 1), seasonal_order=(1, 0, 1, 52))
sarimax_result = sarimax_model.fit()
```

```python
last_date = store_30_data['Date'].max()  # Use the last date for forecasting
forecast_index = pd.date_range(start=last_date, periods=13, freq='W')[1:]  # Periods = 12 weeks

# Use the last 12 weeks of exogenous variables for forecasting
forecast_exog = store_30_data[['Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']].iloc[-12:]

# Get the forecast
forecast = sarimax_result.get_forecast(steps=12, exog=forecast_exog)
forecast_values = forecast.predicted_mean

# Step 9: Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(store_30_data['Date'], store_30_data['Weekly_Sales'], label='Historical Sales')
plt.plot(forecast_index, forecast_values, label='Forecasted Sales', color='red')
plt.title('Sales Forecast for Store 30')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.legend()
plt.show()
```
